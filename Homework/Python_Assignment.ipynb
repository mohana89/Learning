6.Count Number of Odd and Even Elements in a List


Input: lst = [1, 2, 3, 4, 5]
Output: (2, 3)

There are 2 even numbers: 2, 4

There are 3 odd numbers: 1, 3, 5

```python
s=[1,2,3,4,5]
even=[]
odd=[]
ec=0
oc=0
for i in s:
    if i%2==0:
        ec=ec+1
        even.append(i)
    else:
        oc=oc+1
        odd.append(i)
print(f"There are {ec} even numbers, and they are {even}\nThere are {oc} odd numbers, and they are {odd}")
```

7.Maximum difference between two consecutive elements in a list.


Input: lst = [1, 7, 3, 10, 5]
Output: 7

The maximum difference is between 3 and 10 (i.e., |3 - 10| = 7).

Input: lst = [10, 11, 15, 3]
Output: 12

The maximum difference is between 15 and 3 (i.e., |15 - 3| = 12).

```python
#p=[1,7,3,10,5]
p=[10,11,15,3]
fin=0
# consecutive num diff
r=len(p)
for i in range(r):
    j=i+1
    if j<r:
        #print("index values are",i,j)
        c=p[i]
        d=p[j]
        #print("list values are",c,d)
        k=c-d
        k=abs(k)
        if k>fin:
            fin=k
            s,t=c,d
print(f"the largest consecutive number diff is {fin} and numbers are {s},{t}")
```
